### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python

floja = loja['Preço'].sum()
floja2 = loja2['Preço'].sum()
floja3 = loja3['Preço'].sum()
floja4 = loja4['Preço'].sum()

faturamento_lojas = {
    'lojas': [1,2,3,4],
    'faturamento': [floja,floja2,floja3,floja4]
}

faturamento_lojas
```

```python
import matplotlib.pyplot as plt

fatura = plt.bar(faturamento_lojas['lojas'],faturamento_lojas['faturamento'])
plt.xticks([1,2,3,4],['Loja 1','Loja 2','Loja 3','Loja 4'])
plt.ylim(1000000,1600000)
plt.bar_label(fatura)
plt.title('Faturamento das Lojas')
plt.show()
```

  As quatro lojas operam acima de 1,3 milhão em faturamento. A hierarquia é estável: Loja 1 (1.534.509,12), Loja 2 (1.488.459,06), Loja 3 (1.464.025,03) e Loja 4 (1.384.497,58). A Loja 4 é a mais fraca no volume gerado.

  O padrão é homogêneo. Eletrônicos, eletrodomésticos e móveis sustentam mais de 80% das vendas em todas as unidades. Brinquedos, utilidades domésticas e livros permanecem nas posições inferiores em todas elas. Não há vantagem competitiva relevante por categoria em nenhuma loja.

* Avaliações

  A média geral é 4. A Loja 3 lidera com 4,04, seguida pela Loja 2 (4,03), Loja 4 (3,99) e Loja 1 (3,97). A média mais baixa da Loja 1 é compatível com seu volume maior de vendas, que amplia a chance de avaliações negativas. A Loja 4, mesmo com baixo volume, não converte isso em melhor reputação. Esse ponto reduz sua atratividade operacional.

* Produtos mais e menos vendidos

  O trio dominante é estável: celular, TV LED e geladeira. Nos extremos inferiores, aparecem cubo mágico, dinossauro de brinquedo e corda de pular. As quatro lojas reproduzem o mesmo comportamento. O portfólio não diferencia nenhuma unidade.

3.png

* Frete médio

  O frete acompanha o faturamento: Loja 1 (34,69), Loja 2 (33,62), Loja 3 (33,07) e Loja 4 (31,27). A Loja 4, mesmo com frete menor, não converte essa vantagem em vendas maiores. Isso indica fragilidade estrutural e pouca elasticidade comercial.

image.png

 * Conclusão

    As lojas operam com padrões semelhantes de demanda, perfil de produtos e categorias. O único ponto de ruptura está no desempenho inferior da Loja 4: menor faturamento do grupo, avaliação abaixo da média apesar do volume reduzido e nenhuma vantagem real que compense os indicadores fracos. Nem mesmo o frete menor não altera o quadro.
